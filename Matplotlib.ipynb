```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Numbers.csv')

# Read columns
years = data['Year'].tolist()
black_values = data['Black'].tolist()

# Create bar graph
plt.figure(figsize=(10, 6))
plt.bar(years, black_values, color='black')

# Add title and labels
plt.title('Black Column Values Over Years')
plt.xlabel('Years')
plt.ylabel('Black')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Numbers.csv')

# Read columns
total_values = data['Total'].tolist()
years = data['Year'].tolist()

# Create figure and subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Plot histogram
ax1.hist(total_values, bins=20, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
ax1.set_title('Distribution of Total Race Numbers')
ax1.set_xlabel('Total Race Numbers')
ax1.set_ylabel('Frequency')
ax1.grid(True)

# Plot line plot
ax2.plot(years, total_values, color='green', marker='o', linestyle='-')
ax2.set_title('Total Race Numbers Over Years')
ax2.set_xlabel('Years')
ax2.set_ylabel('Total Race Numbers')
ax2.grid(True)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Numbers.csv')

# Filter data for the year 2019
data_2019 = data[data['Year'] == 2019]

# Read columns
columns = ['White', 'Black', 'Hispancic', 'Asain/Native Hawaiian and Pacific Islander', 'American Indian/Alaska Native', 'Multiple Races']

# Get the sum of each column for the year 2019
values = [data_2019[column].sum() for column in columns]

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=columns, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Percentages of Race in Poverty in 2019')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Numbers.csv')

# Read columns 
black_values = data['Black'].tolist()
total_values = data['Total'].tolist()

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(black_values, total_values, color='black', alpha=0.7)

# Add diagonal line
min_black, max_black = min(black_values), max(black_values)
min_total, max_total = min(total_values), max(total_values)
plt.plot([min_black, max_black], [min_total, max_total], color='red', linestyle='--', label='Diagonal Line')

# Add title and labels
plt.title('Total Race Numbers vs. Black Column')
plt.xlabel('Black Column')
plt.ylabel('Total Race Numbers')

# Show legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Numbers.csv')

# Read columns'
years = data['Year'].tolist()
black_values = data['Black'].tolist()
white_values = data['White'].tolist()


# Create stack plot
plt.figure(figsize=(10, 6))
plt.stackplot(years, black_values, white_values, labels=['Black', 'White', ], colors=['black', 'lightgray', 'gray'])

# Add title and labels
plt.title('Race Numbers Over Years (Stack Plot)')
plt.xlabel('Years')
plt.ylabel('Number of People')

# Set y-axis tick locator and formatter for more accurate display
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))

# Change the color of the 'White' column
plt.gca().get_children()[1].set_facecolor('darkorange')

# Show legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
